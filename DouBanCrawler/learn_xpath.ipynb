```python
import requests
from lxml import etree
import pandas as pd
```

```python
url = 'https://movie.douban.com/subject/25934014/comments'
```

```python
r = requests.get(url)
```

```python
s = etree.HTML(r.text)
```

## Copy from browser

```python
#copy from browser, which is not practical,and can only get one single element
print(s.xpath('/html/body/div[3]/div[1]/div/div[1]/div[4]/div[2]/div[2]/p/text()'))
```

```python
#same as above
print(s.xpath('/html/body/div[3]/div[1]/div/div[1]/div[4]/div[4]/div[2]/p/text()'))
```

## Write own Xpath

```python
# much better, can retrieve content in the same position of the whole html
content = s.xpath('//div[@class="comment"]/p/text()')
```
